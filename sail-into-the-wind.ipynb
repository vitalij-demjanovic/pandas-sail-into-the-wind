```python
import pandas as pd

data = pd.read_csv('wind.csv')

data
```

```python
data['Date'] = pd.to_datetime(data[['Yr', 'Mo', 'Dy']].astype(str).agg('-'.join, axis=1), format='%y-%m-%d')
data.drop(columns=['Yr', 'Mo', 'Dy'], inplace=True)
data
```

```python
date_col_index = data.columns.get_loc("Date")
cols = data.columns.tolist()
cols.insert(0, cols.pop(date_col_index))

data = data[cols]

data
```
